```python
names=[]
price=[]
rating=[]
ram=[]
memory=[]
camera=[]
screen=[]
for k in range(1,38):
  url="https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&otracker=nmenu_sub_Electronics_0_Vivo&p%5B%5D=facets.brand%255B%255D%3Dvivo"
  r=requests.get(url)
  htmlContent=r.content
  soup=BeautifulSoup(htmlContent,'html.parser')
  for a in soup.find_all('a',class_="_1fQZEK"):
    t=[]
    for i in a.find_all("li",class_='rgWa7D'):
      t.append(i.get_text())
    memory.append(t[0])
    screen.append(t[1])
    camera.append(t[2])
    try:
      names.append(a.find('div',class_="_4rR01T").get_text())
    except:
      names.append("Nan")
    try:
      price.append(a.find('div',class_="_30jeq3 _1_WHN1").get_text())
    except:
      price.append("Nan")
    try:
      rating.append(a.find('div',class_="_3LWZlK").get_text())
    except:
      rating.append("Nan")
```

```python
import pandas as pd
```

```python
data={"Name":names,"Price":price,"Rating":rating,"Memory":memory,"Camera":camera,"Screen":screen}
df=pd.DataFrame(data)
```
